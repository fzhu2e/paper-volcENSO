```python
import os
import pickle
import numpy as np
import pandas as pd
import xarray as xr
from scipy.signal import find_peaks

import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns

from tqdm import tqdm
import string

import pyleoclim as pyleo
import LMRt
```

## Load data

```python
exp_dict = {
    'LMR (Corals)': '../recons/recon_Corals.nc',
    'LMR (Li13b6)': '../recons/recon_Li13b6.nc',
    'LMR (Corals+Li13b6)': '../Recons/recon_Corals_Li13b6.nc',
}
```

```python
nino34 = {}
year = {}
for exp_tag, path in exp_dict.items():
    with xr.open_dataset(path) as ds:
        nino34[exp_tag] = ds['nino34'].values
        year[exp_tag] = ds['year'].values
```

```python
# SI Table 3 of Toohey et al. (2019): https://doi.org/10.1038/s41561-018-0286-2 + eVolv2k v3 (VSSI>6)
vssi = {
    1107: 19.16,
    1170: 18.05,
    1182: 10.05,
    1191: 8.53,
    1229: 23.78,
    1257: 59.42,
    1275: 11.53,
    1286: 15.06,
    1343: 15.11,
    1452: 9.97,
    1458: 32.98,
    1586: 8.51,
    1594: 8.87,
    1600: 18.95,
    1640: 18.68,
    1695: 15.74,
    1783: 20.81,
    1809: 19.26,
    1815: 28.08,
    1835: 9.48,
    1883: 9.34,
    1991: 8.50,
}

# load Palmyra
df_palmyra = pd.read_table('../data/palmyra2013.txt', skiprows=118)
corals_available_yrs = sorted(list(set([int(t) for t in df_palmyra['age'].values])))

eruption_yrs = sorted(list(vssi.keys()))
print(len(eruption_yrs), eruption_yrs)
```

## SEA

```python
events = [e for e in eruption_yrs if e in corals_available_yrs]
print(len(events), events)

years = np.arange(1, 2002)
preyr, postyr = 3, 6
events_expanded = set()
for e in events:
    idx = list(years).index(e)
    subset = set(years[idx:idx+postyr+1])
    events_expanded |= subset
    
events_expanded = list(events_expanded)
                
start_yr, end_yr = eruption_yrs[0], eruption_yrs[-1]
nonevents = list(np.arange(start_yr-preyr, end_yr+postyr+1))
for e in events_expanded:
    if e in nonevents:
        nonevents.remove(e)
    
# print(len(nonevents), nonevents)

nonevents = [e for e in nonevents if e in corals_available_yrs]
print(len(nonevents), nonevents)
```

```python
# SEA analysis
sea_res = {}
nboot = 1000
for exp_tag in exp_dict.keys():
    volc_yrs = events
    nonvolc_yrs = nonevents
        
    sea_res[exp_tag] = LMRt.utils.sea_dbl(
        year[exp_tag], nino34[exp_tag][2, :], volc_yrs, nonevents=nonvolc_yrs,
        preyr=3, postyr=6, nsample=len(volc_yrs),
        qs=[0.05, 0.5, 0.95], qs_signif=[0.01, 0.05, 0.10, 0.90, 0.95, 0.99],
        draw_mode='non-events', nboot_event=nboot, seeds=np.arange(nboot),
    )
```

## Visualization

```python
from matplotlib.ticker import MultipleLocator

pyleo.set_style('journal', font_scale=1.5)

exp_color = {
    'LMR (Corals)': sns.xkcd_rgb['orange'],
    'LMR (Li13b6)': sns.xkcd_rgb['teal'],
    'LMR (Corals+Li13b6)': sns.xkcd_rgb['pale red'],
}


fig = plt.figure(figsize=[15, 4])
gs = gridspec.GridSpec(1, 3)       
gs.update(wspace=0.3, hspace=0.3)


i = 0
#====================================================
# SEA
#----------------------------------------------------
ax_sea = {}
xlim = [-3.5, 6.5]
ylim = [-1.5, 1.5]
yticks = np.linspace(-1.5, 1.5, 7)
plot_mode = 'composite_norm'
for exp_tag in exp_dict.keys():
    volc_yrs = events
        
    ax_sea[exp_tag] = plt.subplot(gs[i])
    
    ylb = 'Nio 3.4 [K]' if i==0 else None
    ax_sea[exp_tag] = LMRt.visual.plot_sea_res(
        sea_res[exp_tag], ylabel=ylb, color=exp_color[exp_tag],
        font_scale=1.5, ax=ax_sea[exp_tag],  label=f'Composite (n={len(volc_yrs)})',
        xlim=xlim, ylim=ylim, plot_mode=plot_mode, yticks=yticks, signif_fontsize=15,
        indi_style='o', indi_alpha=0.3, xlabel='Years relative to eruption',
    )
    ax_sea[exp_tag].set_title(exp_tag, weight='bold', color=exp_color[exp_tag], y=1.05, fontsize=20)
    ax_sea[exp_tag].legend(frameon=False, loc='lower left', ncol=1, fontsize=15, bbox_to_anchor=(0, -0.5))
    ax_sea[exp_tag].xaxis.set_major_locator(MultipleLocator(2))
    ax_sea[exp_tag].xaxis.set_minor_locator(MultipleLocator(1))
    ax_sea[exp_tag].tick_params(which='minor', length=2, width=1)

    
    i += 1




#====================================================
fs = 20
for i, tag in enumerate(ax_sea.keys()):
    ax_sea[tag].text(
        -0.15, 1.05, string.ascii_lowercase[i],
        transform=ax_sea[tag].transAxes, 
        size=fs, weight='bold',
    )
    
LMRt.showfig(fig)
LMRt.savefig(fig, '../figs/Fig-3.pdf')
```
